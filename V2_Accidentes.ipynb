```python
import pandas as pd

# Cargar el archivo Excel y explorar sus hojas disponibles
archivo = "09_05_01_Accidentes-de-transito-hechos-victimas-y-lesionados-y-hechos-por-zona-de-ocurrencia.xlsx"
xls = pd.ExcelFile(archivo)
print(xls.sheet_names)  # Ver nombres de las hojas disponibles

# Cargar la hoja principal
df = pd.read_excel(xls, sheet_name="total provincial")
df.head(15)  # Ver las primeras filas

```

```python
# Eliminar las primeras 5 filas
df = df.iloc[5:]

df.head(15)  # Ver las primeras filas (ahora sin las 5 primeras originales)
```

```python
# prompt: ver ultimas 10 filas

df.tail(10)
```

```python
# Eliminar las últimas 8 filas
df = df.iloc[:-8]

df.tail(10)
```

```python
# prompt: eliminar columnas completamente vacias

df = df.dropna(axis=1, how='all')
```

```python
#ver el df después de eliminar columnas vacías

df.head()
```

```python
# Eliminar columnas completamente vacías
df = df.dropna(axis=1, how='all')

# Mostrar las primeras filas del DataFrame resultante
df.head()
```

```python
# prompt: cantidad de columnas

# Obtener la cantidad de columnas
num_columnas = len(df.columns)
print(f"La cantidad de columnas en el DataFrame es: {num_columnas}")
```

```python
# Eliminar columnas 4, 8 y 11 (ajustando a índices de Python que empiezan en 0)
columnas_a_eliminar = [3, 7, 10]  # Índices de columnas a eliminar (Python empieza en 0)
df = df.drop(df.columns[columnas_a_eliminar], axis=1)

# Mostrar las primeras filas del DataFrame resultante después de eliminar las columnas
df.head()
```

```python

# Mostrar las primeras filas del DataFrame resultante después de eliminar las columnas
df.head(15)
```

```python
# prompt: cantidad de columnas

# Obtener la cantidad de columnas
num_columnas = len(df.columns)
print(f"La cantidad de columnas en el DataFrame es: {num_columnas}")
```

```python

# Cambiar el nombre de las columnas
nuevas_columnas = ["Año", "Mes", "Total_Accidentes", "Hechos_Simples", "Hechos_Mortales",  "Victimas", "Hechos_Con_Lesiones", "Lesionados",  "Zona_Urbana", "Zona_Rural"]
df.columns = nuevas_columnas[:len(df.columns)] #Asegura que coincida la longitud

# Mostrar las primeras filas del DataFrame resultante con los nuevos nombres de columna
df.head(15)
```

```python
# prompt: Reemplazar todos los valores NaN en la columna "Año" con el valor no nulo más reciente que se encuentra arriba en la misma columna

# Reemplazar valores NaN en la columna "Año" con el valor no nulo más reciente
df['Año'] = df['Año'].fillna(method='ffill')

# Mostrar las primeras filas del DataFrame resultante
df.head(15)
```

```python
# prompt: metricas de las columnas

# Mostrar información estadística de las columnas numéricas
print(df.describe())

# Mostrar el tipo de dato de cada columna
print(df.dtypes)
```

```python
# prompt: convertir las columnas 3 a 10 a datos de tipo numero entero

# Convertir las columnas 3 a 10 a tipo entero (índices de Python)
for i in range(2, 9):  # Rango de columnas 3 a 9 (inclusive)
    try:
        df.iloc[:, i] = pd.to_numeric(df.iloc[:, i], errors='coerce').astype('Int64')
    except KeyError:
        print(f"La columna {i} no existe en el DataFrame.")

# Mostrar el tipo de dato de cada columna después de la conversión
df.dtypes
```

```python
# prompt: convertir columnas Total_Accidentes,Hechos_Simples,Hechos_Mortales,Victimas,Hechos_Con_Lesiones,Lesionados,Zona_Urbana,Zona_Rural a numero entero

# Convertir las columnas especificadas a tipo entero
columnas_a_convertir = ["Total_Accidentes", "Hechos_Simples", "Hechos_Mortales", "Victimas", "Hechos_Con_Lesiones", "Lesionados", "Zona_Urbana", "Zona_Rural"]
for columna in columnas_a_convertir:
    try:
        df[columna] = pd.to_numeric(df[columna], errors='coerce').astype('Int64')
    except KeyError:
        print(f"La columna '{columna}' no existe en el DataFrame.")

# Mostrar el tipo de dato de cada columna después de la conversión
df.dtypes
```

```python
# prompt: mostrar medidas estadisticas de las columnas que ya fueron convertidas a numero entero

# Mostrar medidas estadísticas de las columnas convertidas a entero
columnas_enteras = ["Total_Accidentes", "Hechos_Simples", "Hechos_Mortales", "Victimas", "Hechos_Con_Lesiones", "Lesionados", "Zona_Urbana", "Zona_Rural"]
print(df[columnas_enteras].describe())
```
